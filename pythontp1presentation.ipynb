#### B.3 D'autres types de variables

-En informatique, les **variables** contiennent des valeurs **entières** mais aussi non entières, on dit **de type flottant**. Par exemple:
```python
a=10
b=a/4```
La variable b reçoit la valeur **2.5**, de type flottant.

-Il existe aussi des variables qui contiennent **une chaîne de caractères**.
Par exemple:
```python
mot="bonjour"
print(mot)```

-Enfin, une variable peut prendre que deux valeurs possibles : **True** ou **False**. Une telle variable est appelée **booléen**
